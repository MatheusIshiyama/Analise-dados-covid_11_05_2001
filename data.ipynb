```python
#Importação das bibliotecas
import pandas as pd
from matplotlib import pyplot as plot
```

```python

#github: CSSEGISandData/COVID-19 / url: https://github.com/CSSEGISandData/COVID-19
#Pegar dados de 11/05/2021
#Para pegar os dados mais atualizados, consulte no repositório o arquivo mais atual e substitua a string csvUrl usando o 'raw link'

csvUrl = 'https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_daily_reports/05-11-2021.csv'

data = pd.read_csv(csvUrl)
```

```python
#Linhas e colunas

data.shape
```

```python
#Colunas

data.columns
```

```python
#Filtro apenas dados do Brasil

brazilData = data[data.Country_Region == 'Brazil']
brazilData
```

```python
#Remover colunas

brazilData = brazilData.drop(columns={'FIPS', 'Admin2', 'Lat', 'Long_', 'Combined_Key'})
brazilData
```

```python
#Renomeando colunas para Português

brazilData.rename(columns={'Province_State': 'Província/Estado', 'Country_Region': 'País/Região', 'Last_Update': 'Última atualização', 'Confirmed': 'Confirmado', 'Deaths': 'Mortes', 'Recovered': 'Recuperados', 'Active': 'Ativo','Incident_Rate': 'Média de Casos', 'Case_Fatality_Ratio': 'Média de Casos Fatais' }, inplace=True)

brazilData
```

```python
#Casos totais confirmados (11/05/2021) / 1:1^6 (1 = 1.000.000)

plot.figure(figsize = (50, 10))
plot.bar('Província/Estado', 'Confirmado', data=brazilData)
plot.title('Casos totais confirmado por Estado (1^6 / 1mi)')
plot.xlabel('Província/Estado')
plot.ylabel('Confirmados')
```

```python
#Casos totais de mortes (11/05/2021)

plot.figure(figsize = (50, 10))
plot.bar('Província/Estado', 'Mortes', data=brazilData)
plot.title('Casos Totais de Morte por Estado')
plot.xlabel('Província/Estado')
plot.ylabel('Mortes')
```

```python
#Casos totais de recuperados (11/05/2021) / 1:1^6 (1 = 1.000.000)

plot.figure(figsize = (50, 10))
plot.bar('Província/Estado', 'Recuperados', data=brazilData)
plot.title('Casos Totais Recuperados por Estado (1^6 / 1mi)')
plot.xlabel('Província/Estado')
plot.ylabel('Recuperados')
```

```python
#Medias de casos por região (11/05/2021)

plot.figure(figsize = (50, 10))
plot.bar('Província/Estado', 'Média de Casos', data=brazilData)
plot.title('Média de Casos por Estado')
plot.xlabel('Província/Estado')
plot.ylabel('Média de Casos')
```

```python
#Média de Casos fatais por região (11/05/2021)

plot.figure(figsize = (50, 10))
plot.bar('Província/Estado', 'Média de Casos Fatais', data=brazilData)
plot.title('Média de Casos Fatais por Estado')
plot.xlabel('Província/Estado')
plot.ylabel('Média de Casos Fatais')
```
